```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from rake_nltk import Rake
```

```python
df = pd.read_csv('https://query.data.world/s/625qespqhe5m2rqdffppyw2iouf32z')
df.head()
```

```python
df = df[['Title','Genre', 'Director', 'Actors', 'Description']]
df.tail()
```

```python
#discarding commas bet actors names and getting 4 actors
df['Actors'] = df['Actors'].map(lambda x: x.split(',')[:4])

df['Genre'] = df['Genre'].map(lambda x: x.lower().split(','))

df['Director'] = df['Director'].map(lambda x: x.split(' '))
df.head()
```

```python
for idx, row in df.iterrows():
#     print(idx, row)
    row['Actors'] = [x.lower().replace(" ", '') for x in row['Actors']]
    row['Director'] = ''.join(row['Director']).lower()
```

```python
# Initialize a new column
df['Key_words'] = ""

for index,row in df.iterrows():
    plot = row['Description']
    
    # Initialize rake and discard stop words and punctuations
    r = Rake()
    # Extract keywords from text
    r.extract_keywords_from_text(plot)
    # Getting dictionary with keywords and scores
    key_words_dict_scores = r.get_word_degrees()
    row['Key_words'] = list(key_words_dict_scores.keys())
    
    
```

```python
# Initialize a new column
df['Key_words1'] = ""

for index,row in df.iterrows():
    plot = row['Description']
    
    # Initialize rake and discard stop words and punctuations
    r = Rake()
    # Extract keywords from text
    r.extract_keywords_from_text(plot)
    # Getting dictionary with keywords and scores
    key_words_dict_scores1 = r.get_word_degrees()
    key_words_dict_scores1_sort = sorted(key_words_dict_scores1.items(), key=lambda x: x[1], reverse=True)
    print(key_words_dict_scores1_sort)
    row['Key_words1'] = list(key_words_dict_scores1_sort)
```

```python
# Top 10 key words
df['top10'] = ""
for idx, row in df.iterrows():
#     print(idx, row)
    row['top10'] = [x for x in row['Key_words1'][:10]]
```

```python
df['Bag_of_words'] = ''
columns = ['Genre', 'Director', 'Actors', 'Key_words']
for index, row in df.iterrows():
    words = ''
    for col in columns:
        words += ' '.join(row[col]) + ' '
    row['Bag_of_words'] = words
    
df = df[['Title','Bag_of_words']]
```

```python
df.head()
```

```python
count = CountVectorizer()
count_matrix = count.fit_transform(df['Bag_of_words'])
cosine_sim = cosine_similarity(count_matrix, count_matrix)
print(cosine_sim)
```

```python
# Create a Series of movie titles, so that the series index 
# can match the row and column index of the similarity matrix.
indices = pd.Series(df['Title'])
```

```python
# this function takes in a movie title as input and returns the top 10 recommended (similar) movies

def recommend(title, cosine_sim = cosine_sim):
    recommended_movies = []
    idx = indices[indices == title].index[0]   # to get the index of the movie title matching the input movie
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)   # similarity scores in descending order
    top_10_indices = list(score_series.iloc[1:11].index)   # to get the indices of top 10 most similar movies
    # [1:11] to exclude 0 (index 0 is the input movie itself)
    
    for i in top_10_indices:   # to append the titles of top 10 similar movies to the recommended_movies list
        recommended_movies.append(list(df['Title'])[i])
        
    return recommended_movies
```

```python
# Test the model
recommend('The Avengers')
```
